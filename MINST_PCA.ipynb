**Without using PCA**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import time
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv')
```

```python
df.sample(5)
```

```python
df.shape
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:,0]
```

```python
Y.head()
```

```python
#spiliting the data in train, test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier()
```

```python
#fiting the model with the 784 cols
knn.fit(X_train,y_train)
```

```python
#finding the time duration which it takes for 784 cols
start = time.time()
y_pred = knn.predict(X_test)
print(time.time() - start)
```

```python
#finding the accuracy score
accuracy_score(y_test,y_pred)
```

**Applying PCA**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#scaling the data for PCA
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#importing PCA from sklearn
from sklearn.decomposition import PCA

#creating an instance of PCA
pca = PCA(n_components=200)  # n_components as a parameter in which we pass the components numbers which we want to select
```

```python
#transforming the X_train and X_test
X_train_trf = pca.fit_transform(X_train)
X_test_trf = pca.transform(X_test)
```

```python
X_train_trf.shape
```

```python
knn = KNeighborsClassifier()
```

```python
#fiting the model to transformed data
knn.fit(X_train_trf,y_train)
```

```python
#finding the time duration for 200 cols
start = time.time()
y_pred = knn.predict(X_test_trf)
print(time.time() - start)
```

```python
#finding accuracy score after PCA
accuracy_score(y_test,y_pred)
```

```python
# A loop which will runs 784 times and will find accuracy score after every iteration from this we can find out that how many features are important to acheive good accuracy score
for i in range(1,785):
    pca = PCA(n_components=i)
    X_train_trf = pca.fit_transform(X_train)
    X_test_trf = pca.transform(X_test)

    knn = KNeighborsClassifier()

    knn.fit(X_train_trf,y_train)

    y_pred = knn.predict(X_test_trf)

    print(accuracy_score(y_test,y_pred))
```
